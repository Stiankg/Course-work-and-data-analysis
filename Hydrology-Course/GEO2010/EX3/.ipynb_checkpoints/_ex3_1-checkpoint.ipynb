1. Atmospheric Moisture:
    Vapour pressure is usually denoted by 'e' and is expressed usually in millibar (mb) or kilopascal (kPa). 1mb =
    100N/m2 and 1mm of Hg = 1.33 mb. The mean sea level standard atmospheric pressure at 15oc is 1013 mb.
    
    The saturation vapour pressure is function of temperature (T). Since it is related to dry bulb air temperature, it can be determine by following equation (Dingman, pp86).
    
     ## $e_a = 6.108 exp\left ( \frac{17.27 \times  T}{T+237.3} \right )$
   
   T = Air temperature in degree centigrade
   
   where, e_a = saturation vapour pressure [mb]
   
   
 Actual Vapour Pressure:
 
  ## $ e_d = e_{a[t-wet]} - \gamma ( T_{dry} - T_{wet}) P $
  
  where, $ \gamma $
  
                  = 0.00066 for assman aspiration at 5m/sec
  
                  = 0.008 for natural ventilation at 1 m/s
                  
                  = 0.0012 for indoor ventilation at 0m/s 
                  
  $ e_{a[t-wet]} $ = Saturation vapor pressure at wet bulb temperature [kPa]
  
  P = atmospheric pressure [kPa]
  
  The atmospheric pressure (P) can be calculated from the altitude (z) of the station with the following equation.
  
  ## $ P = 101.3 \left ( \frac{293-0.0065\times z}{293} \right )^{5.256}$
  
  Where, Z = elevation of station above sea level (m)
  
 Relative Humidity:
 
 it can be calculated by : 
 
 ### $ RH   = \frac{e_d}{e_a}$
 
 [Note: $e_d = ea $ at dew point temperature.]
 
 **Alternatively $e_d$ can be calculated by:**
 
 
 ## $e_d = 0.611 \times exp \left (\frac{ 17.3 \times T_d}{T_d + 237.3}  \right )$

**Some defination:**

Absolute humidity is the measure of water vapor (moisture) in the air, regardless of temperature. It is expressed as grams of moisture per cubic meter of air $(g/m^3)$.

Relative humidity also measures water vapor but RELATIVE to the temperature of the air. It is expressed as the amount of water vapor in the air as a percentage of the total amount that could be held at its current temperature.

saturation vapor pressure: This is a maximum possible vapor pressure at at given temperature at which vapor starts to condense. It is measured in [mbar/kPa].

Actual vapor pressure: This is pressure exerted by water vapor actually contained in the air. It is measured in [mbar/kPa].

Vapor pressure deficit: this is difference between saturation and actual vapour pressure. It is measured in [mbar/kPa].

Dew point temperature: This refers to temperature to which air needs to be cooled to become saturated. It is measured in $(^oK/^oC)$
