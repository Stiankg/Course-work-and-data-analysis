```python
from openpyxl import *
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

#Open workbook and worksheet
wb = load_workbook(filename ='HauglandLiavatn.xlsx')
ws = wb.active

#Extract the data from the worksheet.
Liavatn  = [float(cell.value) for row in ws.iter_rows('B3:B9498') for cell in row]
Haugland = [float(cell.value) for row in ws.iter_rows('C3:C9498') for cell in row]
Weibull  = [float(cell.value) for row in ws.iter_rows('D3:D9498') for cell in row] 

#Convert lists to numpy arrays.
Liavatn = np.array(Liavatn); Haugland = np.array(Haugland); Weibull = np.array(Weibull)

#Now we should find some neat way to sort the data.
#This sorts the data in a descending order.
Liavatn[::-1].sort(); Haugland[::-1].sort()

N = len(Weibull)
Weibull = (1.0/(N+1))*Weibull #Convert Weibull to actual Weibull numbers.
```

```python
plt.figure(1)
plt.plot(Weibull,Liavatn,'-r',label='Liavatn')
plt.plot(Weibull,Haugland,'-b',label='Haugland')
plt.legend(loc='upper right')
plt.ylabel('sorted data',fontsize=12)
plt.xlabel('1-F(i)',fontsize=12)
plt.show()
```
