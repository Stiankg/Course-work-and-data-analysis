## Exercise 10.1: The Linear-Reservoir Model

**<font color='blue'>
Reference material: Dingman 3rd ed., p.471-77
</font>**

**Preliminary question:** What is the Linear-Reservoir Model and what is it used for? 

**Theory**

The discharge $q^*$ at time $t$ is defined as follows

$$q^*(t) =
\begin{cases}
p^* \cdot (1- \exp(-t/T^*)), & \quad t \leq T_{p}\\
q_{pk}^* \cdot \exp(-(t-T_{p})/T^*), & \quad t \geq T_{p}\\
\end{cases}$$

where

* $q^*$ is discharge 	[L T<sup>-3</sup>] or [L T<sup>-1</sup>] <br />
* $q_{pk}^*$ is peak discharge 	[L T<sup>-3</sup>] or [L T<sup>-1</sup>] <br />
* $p^*$ is efficient precipitation rate	 [L T<sup>-3</sup>] or [L T<sup>-1</sup>] <br />
* $t$ is time <br />
* $T_{p}$ is input duration <br />
* $T^*$ is response time <br /> </p>
   
**<font color='blue'>
See also Dingman 3rd ed., p.472, Box 10.3; p.505, equation 10.52
</font>**

a) Use the Linear-Reservoir Model to calculate the discharge in response to these inputs: $p^*$, $T^*$. 
Plot the first 20 hours of discharge at half-hour resolution with the following values:

_i)_	$p^*$ = 1 unit/hour, $T^*$ = 1 hour, $T_{p}$ = 5 hours <br />
_ii)_	$p^*$ = 1 unit/hour, $T^*$ = 2 hours, $T_{p}$ = 5 hours

Is 20 hours enough to describe the whole event? How can one interpret $T^*$?

<font color='green'>

**_A few hints to implement the Linear-Reservoir Model_** <br />
<br />
What we know: <br />
<br />
* $dt$ = 0.5 hours, model timestep <br />
* $T$ = 20 hours, length at which the model is to be run <br />
<br />
What we need to calculate: <br />
 $N = T/dt + 1$ total number of timesteps at which the model is to be run <br />
<br />
Steps: <br />
- Definition of a time vector $t$ of dimension $N$ <br />
- Loop on time vector <br />
- Conditional loops on time vector to calculate discharge $q^*(t)$

</font>

```python
#Import libraries:
import numpy as np
import matplotlib.pyplot as plt

#Declare constants:
dt = 0.5    #time step [h]
Tend = 20   #end of time [h]

#Create array of time 
N = int(Tend/dt)+1    #number of time steps
t = np.linspace(0,Tend,N )

#Create an empty array for q:
q=np.zeros(N)


def calc_and_plot(precip, Tstar, Tpeak, plotcolor):
    """
    This function calculates the outflow and plots it as a line.
    """
    i=0
    time=t[i]
    #iterate through the input time and calculate the outflow:
    while time  <= Tpeak:
        q[i] = precip*(1-np.exp(-time/Tstar))
        i = i+1 
        time = t[i]

    #Set q_peak equal to the outflow at the end of the input phase
    q_peak=q[i-1]

    #iterate through the remaining time steps and calculate outflow:
    while time < Tend:
        q[i] = q_peak*np.exp(-(time-Tpeak)/Tstar)
        i=i+1
        time = t[i]
    #Plot the outflow as a function of time:    
    plt.plot(t,q, c = plotcolor,label = 'p$^*$=' +str(precip)+ ' unit/h, T$^*$ = ' +str(Tstar) + ' h')
    return

#Call the function for each set of parameters:
T_peak = 5
calc_and_plot(1, 1, T_peak, 'blue')
calc_and_plot(1, 2, T_peak, 'red')


plt.title('Linear-Reservoir Model outflow with $T_p$ = ' + str(T_peak))
plt.xlabel('Time [h]')
plt.ylabel('Outflow [unit/h]')
plt.legend(loc='upper right')
plt.show()
```

b) Now do the same exercise with different values for $q_{pk}^*$, to see how the discharge in the Linear-Reservoir Model responds to different parameters. Choose three values of $q_{pk}^*$ (justify the choice) and show three figures with two curves each ($T^*$ = 1 and $T^*$ = 2). Comment your findings.

```python
#Make plots for different values of q_peak, i.e. for different rates of precipitation. 
# We choose p = 5, 10 and 15, to leave some time before and after the peak

plt.figure(1)

precip = 1

plt.subplot(311)
calc_and_plot(precip, 1, T_peak, 'blue')
calc_and_plot(precip, 2, T_peak, 'red')
plt.title('Precipitation rate =' + str(precip)+' unit/h')
plt.legend(loc='upper right')
plt.xlabel('Time [h]')
plt.ylabel('Outflow [unit/h]')

precip = 5
plt.subplot(312)
calc_and_plot(precip, 1, T_peak, 'blue')
calc_and_plot(precip, 2, T_peak, 'red')
plt.title('Precipitation rate =' + str(precip)+' unit/h')
plt.legend(loc='upper right')
plt.xlabel('Time [h]')
plt.ylabel('Outflow [unit/h]')

precip = 10
plt.subplot(313)
calc_and_plot(precip, 1, T_peak, 'blue')
calc_and_plot(precip, 2, T_peak, 'red')
plt.title('T_peak =' + str(T_peak))
plt.title('Precipitation rate =' + str(precip)+' unit/h')
plt.xlabel('Time [h]')
plt.ylabel('Outflow [unit/h]')
plt.legend(loc='upper right')
plt.tight_layout()

```

## Exercise 10.2: The Unit Hydrograph

**<font color='blue'>
Reference material: Dingman 3rd ed., p.521-29
</font>**

**Preliminary question:** What is the Linear-Reservoir Model and what is it used for? 

Note: Abbreviation UH stands for Unity Hydrograph.

In this exercise we have two Excel files, *UH.xlsx* and *Baseflow.xlsx*. *UH.xlsx* contains the time [hr] in the UH in column (A2:A12) and the UH [m<sup>3</sup> s<sup>-1</sup>] in column (B2:B12). *Baseflow.xlsx* contains the time [hr] in the baseflow in column (A2:A17) and the baseflow [m<sup>3</sup> s<sup>-1</sup>] in column (B2:B17). 

a) Read the data from the Excel files into list and give them meaningful names. Then convert the lists to numpy arrays for numerical manipulations.

```python
#These import statements is all you need to solve the exercise.
from openpyxl import *
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = (10.0, 8.0) #Set size of inline plot

#a)
# Define a function that reads the prescibed data:
def read_xls_colRange(filename,worksheet_no, rowStart, rowEnd, col):
    import xlrd
    """
    This function reads a range of cells specified by ROWSTART, ROWEND and
    column number COL (column A = 0, column B = 1 etc) in worksheet number 
    WORKSHEET_NO in file FILENAME.
    """    
    #Open the file:
    file = xlrd.open_workbook(filename) 
    #Open the work sheet
    ws = file.sheet_by_index(worksheet_no)
    #Declare an empty range that we will fill and then return:
    N = rowEnd-rowStart+1
    cellVal = np.zeros(N)
    rowRange = range(rowStart,rowEnd+1)
    #Read the data:
    for row in rowRange:
            cellVal[row-rowStart]=ws.cell(row,col).value
    return cellVal

baseflowTime = read_xls_colRange('Baseflow.xlsx', 0, 1, 16, 0)
baseflow = read_xls_colRange('Baseflow.xlsx', 0, 1, 16, 1)
UHtime = read_xls_colRange('UH.xlsx', 0, 1, 11, 0)
UHflow = read_xls_colRange('UH.xlsx', 0, 1, 11, 1)

```

b) Rainfall amounts are provided for three periods of 3 hours each. Assume 20mm rain infiltrates the ground in each of these periods. Calculate the effective input for each period. Then calculate the discharge every 3 hours up to 42 hours after the start of the following rainfall periods 

* Period 1: 30mm rainfall between 0600-0900  <br />
* Period 2: 45mm rainfall between 0900-1200  <br />
* Period 3: 50mm rainfall between 1800-2100  <br />

Let our time series be represented by the baseflow series and treat each rainfall period individually using the UH.

Our units are as follows: One unit of time is 3 hours, one unit of rain is 1 mm and one unit of outflow is m<sup>3</sup>/s.

```python
#b)
#Calculate effective inputs [mm]:
effInputP1 = 30 - 20    #
effInputP2 = 45 - 20
effInputP3 = 50 - 20

#Expand the event flow arrays with zeros so that it gets the same size
#as baseflow:

UHlong=np.append(UHflow, np.zeros(len(baseflow)-len(UHflow)))

#Calculate outflow (Q) for each period:
eventflow1 = effInputP1*UHlong
eventflow2 = effInputP2*UHlong
eventflow3 = effInputP3*UHlong

#Superimpose the event flows on the baseflow:
Q1=eventflow1+baseflow
Q2=eventflow2+baseflow
Q3=eventflow3+baseflow

#Plot the event flows:

plt.plot(baseflowTime,Q1, label = 'Period 1')
plt.plot(baseflowTime,Q2, label = 'Period 2')
plt.plot(baseflowTime,Q3, label = 'Period 3')
plt.plot(baseflowTime,baseflow, label = 'Baseflow')

plt.xlabel('time, [h]')
plt.ylabel('Outflow, [$m^3/s$]')
plt.title('Non-synchronous unit hydrographs')
plt.legend()

plt.show()
```

c) Find the indices $s_1, s_2$ and $s_3$ corresponding to the starting time of each rainfall period in the time vector.
Use the indices to synchronize the discharge vector of each period with the baseflow values. Compute both the event flow and the total flow (i.e. total discharge). <br />
<br />
<font color='green'>
Note that each hydrograph vector has size(UH,1) entries, i.e. [size(UH,1)-1] 3 hour periods. 
</font>

```python
#c)
#Declare the start times:
s1time = 6
s2time = 9
s3time = 18

#Find the indices of the starting times in the baseflow time array:
for i in range(0,len(baseflowTime)):
    if baseflowTime[i]==s1time:
        s1i = i
    if baseflowTime[i]==s2time:
        s2i = i
    if baseflowTime[i]==s3time:
        s3i = i
        break    #abbort the for-loop when the last index has been found

#Add zeros to the beginnings and ends of the eventflows to
#make them the same size as baseflow:
event1 = np.zeros(len(baseflow))
event2 = np.zeros(len(baseflow))
event3 = np.zeros(len(baseflow))

for i in range(0,len(baseflow)):
    """
    Iterate through the long arrays and add the flows at the right times
    """
    if i>s1i and i<s1i+len(eventflow1):
        event1[i]=eventflow1[i-s1i]
    
    if i>s2i and i<s2i+len(eventflow2):
        event2[i]=eventflow2[i-s2i]
    
    if i>s3i and i<s3i+len(eventflow3):
        event3[i]=eventflow3[i-s3i]
        
#Add the eventflows and the baseflow:
total = event1+event2+event3+baseflow
```


d) Finally plot the results, with the discharge from the 10mm, 25mm and 30mm, hydrographs the baseflow, the event flow and the total discharge. Try to make the plot intelligible in the sense that you include axis labels, title and legends.

```python
#d) 
plt.plot(baseflowTime,total, label='Total')
plt.plot(baseflowTime, event1, label='Period 1')
plt.plot(baseflowTime, event2, label='Period 2')
plt.plot(baseflowTime, event3, label='Period 3')
plt.plot(baseflowTime, baseflow, label='Period 3')

plt.xlabel('time, [h]')
plt.ylabel('Outflow, [$m^3/s$]')
plt.title('Synchronized hydrographs')
plt.legend()

plt.show()
```
