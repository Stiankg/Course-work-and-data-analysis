```python
#Solution of Exercise 10.2: The Unit Hydrograph

from openpyxl import *
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt



wb = load_workbook('UH.xlsx')
ws = wb.active

t1 = [float(cell.value) for row in ws.iter_rows('A2:A12') for cell in row] #Time in the UH [hours]
UH = [float(cell.value) for row in ws.iter_rows('B2:B12') for cell in row] #Unity hydrograph [m3 s-1]

wb = load_workbook('Baseflow.xlsx')
ws = wb.active

t2 = [float(cell.value) for row in ws.iter_rows('A2:A17') for cell in row] #Time in the baseflow [hours]
Base = [float(cell.value) for row in ws.iter_rows('B2:B17') for cell in row]

#Convert lists to numpy arrays
Base = np.array(Base); UH = np.array(UH); t1 = np.array(t1); t2 = np.array(t2)
```

```python
#b) Calculate the discharge(?) every 3 hours up to 42 hours after the start of the following rainfall periods: 
#30mm 0600-0900, 45mm 0900-1200 and 50mm 1800-2100. Assume that 20mm rain infiltrates the ground in each period.

#First we calculate the effective input in each period:
R1=30; R2=45; R3=50 #Rainfall in each period [mm/3h]
infil=20 #Infiltration [mm/3h]
W1=R1-infil; W2=R2-infil; W3=R3-infil #Effective input [mm/3h]

#Let our time series be represented by the baseflow series
t = t2

#Treat each rainfall period individually using the UH:
Q1h = W1*UH
Q2h = W2*UH
Q3h = W3*UH
```

```python
#Now we have to synchronize these series with the baseflow:
Q1 = np.zeros(len(t))
Q2 = np.zeros(len(t))
Q3 = np.zeros(len(t))

#Find indices corresponding to the starting time of each rainfall period in the time vector:
s1 = 0; s2 = 0; s3 = 0

index = 0
while t[index] != 6:
    index+=1
    s1 = index

index = 0
while t[index] != 9:
    index+=1
    s2 = index
    

index = 0
while t[index] != 18:
    index+=1
    s3 = index
```

```python
#Synchronize the discharge vectors for each period with the baseflow.
#Each hydrograph vector has size(UH,1) entries,  i.e. [size(UH,1)-1] 3 hour periods:

lenhyd = len(UH)
Q1[s1:s1+lenhyd] = Q1h
Q2[s2:s2+lenhyd] = Q2h

#The last precipitation event runs past 45 hours:
Q3[s3:] = Q3h[0:(len(Q3)-s3)]
```

```python
#Compute both the event flow and the total flow (i.e. total discharge):
eventflow = Q1+Q2+Q3
totalflow = Q1+Q2+Q3+Base
```

```python
#Plot the results, with the discharge from the 10mm, 25mm and 30mm hydrographs, the baseflow, the event flow
#and the total discharge:

matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)

plt.figure(1)
plt.plot(t,totalflow,label='Total discharge')
plt.plot(t,eventflow,label='Event flow')
plt.plot(t,Q1,'--',label='Hydrograph 10')
plt.plot(t,Q2,'--',label='Hydrograph 25')
plt.plot(t,Q3,'--',label='Hydrograph 30')
plt.plot(t,Base,'--',label='Baseflow')
plt.legend(loc='upper right')
plt.xlabel('Time (hours), 3h resolution',fontsize=12)
plt.ylabel('Discharge, $Q_i$ (m^3/s)',fontsize=12)
plt.title('Discharge using the Unit Hydrograph, $Q_i(t)$',fontsize=12)
plt.show()
```
